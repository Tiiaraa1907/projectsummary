#1. Missing Value in Gender Column

```python
# missing value in Gender column
csmiss= cs['Gender'].isna().sum(axis=0)
print(csmiss)

gendercount=cs['Gender'].count()
print(gendercount)

gendercalc=csmiss/gendercount*100
print("Missing value in gender column : ",gendercalc,"%")
```

#2. Find which column that has the outlier

```python
numcols = ['id','Region_Code','Annual_Premium','Policy_Sales_Channel','Vintage']

fig, axes = plt.subplots(2,4, figsize=(32,18))
i = 0
j = 0
for col in numcols:
    sns.boxplot(x=col, ax=axes[i,j], data=cs)
    sns.despine()
    if j != 3:
        j += 1
    else:
        i += 1
        j = 0

fig.suptitle('Boxplots Case', fontsize=24)

# Show plot
plt.show()
```

#5. The dominan category in vehicle_age column

```python
# the dominan category in vehicle_age column
# Set up the size of canvas
fig, ax = plt.subplots(1,1,figsize=(10,5))
# Hide the top and right axis lines
sns.despine()

# Histogram plot of Umur (x for horizontal, y for vertical)
sns.histplot(x=cs['Vehicle_Age'])

# Add title
ax.set_title('Histogram of Vehicle_Age', fontsize=14)

# Show plot
plt.show()
```
