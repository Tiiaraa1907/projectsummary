```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

# 1. Count of one categorical feature

```python
df1 = df['MSSubClass'].value_counts()

# Create a bar plot
sns.barplot(x=df1.index, y=df1.values)
plt.xlabel('MSSubClass')
plt.ylabel('Count ID')
plt.show()
```

# 2. Distribution of one continuous feature

```python
sns.displot(df['SalePrice'],kind='hist',kde=True)
```

# 3. Categorical-continuous

```python
fig = plt.figure()
continuous_feature = sns.boxplot(data = df, x='MSSubClass', y = 'MoSold')
plt.show()
```
